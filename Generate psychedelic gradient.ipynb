```python
from PIL import Image
import random, math
from IPython.display import display  # for Colab/Jupyter

# Image size
w, h = 300, 300;img = Image.new("RGB", (w, h))


# Generate psychedelic gradient
for y in range(h):
    for x in range(w):
        sin_component = 127 * math.sin(x / 15)
        cos_component = 127 * math.cos(y / 15)
        noise = random.randint(0, 50)
        v = int(sin_component * cos_component + noise)
        img.putpixel((x, y), (v % 256, (v * 2) % 256, (v * 3) % 256))

# Show image (works in Colab/Jupyter)
display(img)

# Save image
img.save("psychedelic_art.png")
```

```python
from PIL import Image
import random, math

# Image size
w, h = 400, 400
img = Image.new("RGB", (w, h))

# Generate psychedelic gradient
for y in range(h):
    for x in range(w):
        # Sin & Cos wave interference
        sin_component = 127 * math.sin(x / 15)
        cos_component = 127 * math.cos(y / 15)

        # Random noise for trippy effect
        noise = random.randint(0, 50)

        # Combine values
        v = int(sin_component * cos_component + noise)

        # Map to RGB
        img.putpixel((x, y), (v % 256, (v * 2) % 256, (v * 3) % 256))

# Show result (opens in default viewer if local)
img.show()

# Save output
img.save("psychedelic_gradient.png")

print("🎨 Psychedelic Gradient Art saved as 'psychedelic_gradient.png'")
```
